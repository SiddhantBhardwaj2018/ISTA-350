## Brief aside: the star operator, `*`
* Sometimes called the `splat` operator
* Functions as a `pack`/`unpack` operator

<font color=red>***In the `SIGNATURE` of a function definition, we use the star operator to `PACK` an unknown number of operators into a tuple. This means that the function definition doesn't need to know how many arguments there will be in advance. Call this parameter `star-args`.***

```python
def sum_squares(*args):
    print(type(args))
    total = 0
    for x in args:
        total += x * x
    return total
```

```python
sum_squares(3, 2, 5)
```

```python
sum_squares(3, 2, 5, 6)
```

<font color=red>***In a `FUNCTION CALL`, we use the star operator to `UNPACK` a sequence and turn its elements into individual arguments:***

```python
import math

def hypotenuse(a, b):
    return math.sqrt(a * a + b * b)
```

```python
sides = [3, 4]
hypotenuse(*sides)
```

<font color=red>***Looks like by default that Python divided the data points into 10 bins, but we wanted our bins to be 0.001 wide, so let's figure out how many bins that would be:***

```python
num_bins = round(max(densities), 3) * 1000 # a float
num_bins -= round(min(densities), 3) * 1000 + 1
```

<font color=red>***Above we used `pandas` `plot` method, which is built on top of the `matplotlib` `pyplot` function. Let's use that this time. Let's pass in the keyword arg `density=1`, which will convert our frequencies into probability densities, which we will talk about later:***

```python
n, bins, patches = plt.hist(densities.values, 
        bins=int(num_bins), density=1, facecolor='green')
```

<font color=blue>***Back to our original question - does this look like a normal distribution to you? <font color=red>If you recall, a normal distribution is completely characterized by its mean and its standard deviation. So let's plot a normal distribution that has the same mean and std on top of the histogram:***

```python
x = np.linspace(0.1, 0.8, 100)
y = norm.pdf(x, loc=densities.mean(), 
                scale=densities.std(ddof=1))
plt.hist(densities.values, bins=int(num_bins), 
         density=1, facecolor='green')
plt.plot(x, y, color='r', linewidth=4)
plt.xlabel('Nest Densities (nests / $m^2$)')
plt.ylabel('Probability Densities')
```

<font color=red>***Clearly not even close.***

<font color=red>***We can also use statistical tests to get a quantitative characterization of how close to normal our distribution is:***

```python
normaltest(densities)
```

<font color=red>***The first number expresses how far away from normal the distribution is. `0` is perfectly normal, so this says it's really far from normal. The p-value is basically 0, so we can believe this result.***

<font color=red>***Recall from 131 $R^2$, a measure of goodness of fit of data to a model, from https://medium.com/@erika.dauria/looking-at-r-squared-721252709098:***
    

<font color=red>***Generally, $R^2$ can be [0, 1], with 0 being totally terrible and 1 being and exact fit. Let's calculate it for this problem:***

```python
def r_squared(data_series, model):
    # normally between 0 and 1
    ybar = data_series.mean()
    numerator = denominator = 0
    for i in range(len(data_series)):
        numerator += (data_series.iloc[i] - model[i])**2
        denominator += (data_series.iloc[i] - ybar)**2  # (y - ybar)**2
    return 1 - numerator / denominator
```

```python
model = norm.pdf(densities.index.values, loc=densities.mean(), 
                scale=densities.std(ddof=1))
        
r_squared(densities, model)
```
